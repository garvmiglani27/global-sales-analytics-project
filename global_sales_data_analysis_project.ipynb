```python
cursor = conn.cursor()
```

```python
cursor.execute("CREATE DATABASE IF NOT EXISTS sales_project")

print("Database created")
```

```python
cursor.execute("USE sales_project")
```

```python
create_table = """
CREATE TABLE IF NOT EXISTS sales (
Order_ID INT,
Order_Date DATE,
Country VARCHAR(100),
Product VARCHAR(100),
Category VARCHAR(100),
Quantity INT,
Unit_Price FLOAT,
Revenue FLOAT
)
"""

cursor.execute(create_table)

print("Table created")
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"], dayfirst=True)
```

```python
for i, row in df.iterrows():
    
    sql = """
    INSERT INTO sales
    (Order_ID, Order_Date, Country, Product, Category, Quantity, Unit_Price, Revenue)
    VALUES (%s,%s,%s,%s,%s,%s,%s,%s)
    """
    
    cursor.execute(sql, tuple(row))
```

```python
conn.commit()

print("Data inserted successfully")
```
